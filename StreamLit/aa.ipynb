```python
import pandas as pd

# فایل اکسل را بارگذاری کنید
input_file = r"C:\Users\s_alizadehnia\Downloads\Car_Dena_TC+5_26 s 513 ir 10_Data_09_01_2024, 05_57_00_to_01_05_2025, 14_58_00.xlsx"  # نام فایل ورودی اکسل
output_file = r"C:\Users\s_alizadehnia\Downloads\Car_Dena_TC+5_26 s 513 ir 10_Data_09_01_2024, 05_57_00_to_01_05_2025, 14_58_00_n.xlsx"  # نام فایل خروجی اکسل


trips_to_keep0 = [
    10, 11, 12, 13, 19, 20,
    21, 22, 23, 24, 27, 28,
    30, 31,  41 ,  177, 179,
    180, 181
]

# لیستی از tripهایی که می‌خواهید نگه دارید
trips_to_keep1 = [
    "trip_10", "trip_11", "trip_12", "trip_13", "trip_19", "trip_20",
    "trip_21", "trip_22", "trip_23", "trip_24", "trip_27", "trip_28",
    "trip_30", "trip_31",  "trip_41", 
    "trip_177", "trip_179",
    "trip_180", "trip_181"
]

# داده‌ها را از اکسل بخوانید
sheets = pd.read_excel(input_file, sheet_name=None)  # خواندن تمام شیت‌ها به صورت دیکشنری

# فیلتر کردن داده‌های شیت "Data"
df_data = sheets.get("Data", pd.DataFrame())
df_data_filtered = df_data[df_data['trip'].isin(trips_to_keep0)]

# فیلتر کردن داده‌های شیت "Report"
df_report = sheets.get("Report", pd.DataFrame())
df_report_filtered = df_report[df_report['report_trip_number'].isin(trips_to_keep1)]

# ذخیره فایل خروجی با شیت‌های جداگانه با داده‌های فیلتر شده
with pd.ExcelWriter(output_file) as writer:
    df_data_filtered.to_excel(writer, sheet_name="Data", index=False)
    df_report_filtered.to_excel(writer, sheet_name="Report", index=False)

print(f"فایل فیلتر شده ذخیره شد: {output_file}")

```

```python
import matplotlib.pyplot as plt
import pickle

# Load the data from the provided pickle file
file_path = r"C:\Users\s_alizadehnia\Desktop\result_230_472_251_106.pkl"  # نام فایل .pkl خود را اینجا قرار دهید
with open(file_path, 'rb') as file:
    data = pickle.load(file)

# Extract the main 'scores' list
scores = data.get('scores', [])

# Plot the scores
plt.figure(figsize=(12, 6))
plt.plot(range(len(scores)), scores, label='Scores', color='blue', marker='o', markersize=4, linewidth=1)

# Add labels, title, and legend
plt.title('Scores Distribution', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Score Value', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
